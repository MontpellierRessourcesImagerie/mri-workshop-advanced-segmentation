```python
import numpy as np
import pandas as pd
```

```python
file = "/home/cedric/Documents/ML/Demos/svm/my_table.csv"
data = pd.read_csv(file)

data.head(n = 5)
```

```python
X = data.drop(columns=['Label', 'Metadata_Well'])
y = data['Label']

print(X)
```

```python
from sklearn.model_selection import train_test_split  

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.svm import SVC

svclassifier = SVC(kernel='linear')  
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
W = data[['ImageNumber', 'ObjectNumber', 'Metadata_Well']]
X_newtest = pd.merge(X_test, W, left_index=True, right_index=True)
X_newtest = X_newtest[['ImageNumber_x', 'ObjectNumber_x', 'Metadata_Well']]
new_testdata = X_newtest.assign(Prediction = y_pred)
print(new_testdata)



from sklearn.metrics import confusion_matrix  
print(confusion_matrix(y_test, y_pred))
```
